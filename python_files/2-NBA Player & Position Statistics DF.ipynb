```python
# Initialize dataframe with required columns
player_df = player_df_sample[0:0]

# List of players
search_player = [236, 553, 402]

# Set url for API
url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

# Loop through each player ID
for player in search_player:
    try:
        # Create search query, make request for player stats and store in json
        
        querystring = {"id":{player},"season":"2021"}

        headers = {
            "X-RapidAPI-Key": "8c034ba04bmsha0b2a7a1ae34c43p11d80ajsneb719a3861d5",
            "X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
        }

        response = requests.get(url, headers=headers, params=querystring)
        response_json=response.json()
        # pprint(response_json)
        
        # Access Player statistics and create a dataframe to append to previous player response
        df=pd.DataFrame(response_json['response'])
        player_df = pd.concat([player_df,df])
        
    # Handle exceptions for a player that is not available in the Pokemon API
    except Exception as e:
        print(f"Error occurred with {player}: {e}")
        pass
    
player_df.columns
```

```python
player_df.head()
```

```python
player_df.reset_index(drop=True,inplace=True)

player_df_player=pd.DataFrame(player_df.loc[:,'player'].values.tolist()).rename(columns={'id':'player_id'})
player_df_player.reset_index(drop=True,inplace=True)
player_df_player

player_df_game=pd.DataFrame(player_df.loc[:,'game'].values.tolist()).rename(columns={'id':'game_id'})
player_df_game.reset_index(drop=True,inplace=True)
player_df_game

player_df_team=pd.DataFrame(player_df.loc[:,'team'].values.tolist()).rename(columns={'id':'team_id','name':'team_name','nickname':'team_nickname'})
player_df_team.reset_index(drop=True,inplace=True)
player_df_team
```

```python
player_df_details=player_df[0:0]
player_df_details = pd.concat([player_df,player_df_player,player_df_game,player_df_team,],axis=1)
player_df_details.drop(columns=['player', 'team', 'game'],axis=1,inplace=True)
player_df_details['min'].fillna('0:0',inplace=True)
player_df_details['min']=player_df_details['min'].astype('str')
player_df_details['tpp'].fillna(0,inplace=True)
player_df_details['tpp']=player_df_details['tpp'].astype('float64')

def convert_to_decimal(time_str):
    minutes, seconds = map(int, time_str.split(":"))
    decimal_time = minutes + (seconds/60)
    return decimal_time

player_df_details['minutes'] = list(map(convert_to_decimal, player_df_details['min']))

player_df_details=player_df_details[['player_id',\
       'firstname', 'lastname', 'team_id', 'team_name', 'team_nickname',\
       'code', 'logo', 'game_id','points', 'pos', 'minutes', 'fgm', 'fga', 'fgp', 'ftm', 'fta', 'ftp', 'tpm',\
       'tpa', 'tpp', 'offReb', 'defReb', 'totReb', 'assists', 'pFouls',\
       'steals', 'turnovers', 'blocks', 'plusMinus', 'comment']]

player_df_details.columns
```

```python
player_df_details.head()
```

```python
player_df_details_1 = player_df_details[['player_id','firstname', 'lastname', 
       'points', 'minutes', 'fgm', 'fga', 'ftm', 'fta', \
       'tpm','tpa', 'tpp', 'offReb', 'defReb', 'totReb', \
       'assists', 'pFouls','steals', 'turnovers', 'blocks', 'plusMinus']]

player_df_aggregate = player_df_details_1.groupby(['player_id','firstname', 'lastname']).agg('sum')
player_df_aggregate['fgp']=player_df_aggregate['fgm']/player_df_aggregate['fga']
player_df_aggregate['ftp']=player_df_aggregate['ftm']/player_df_aggregate['fta']
player_df_aggregate.columns
```

```python
player_df_details_2 = player_df_details[['pos',
       'points', 'minutes', 'fgm', 'fga', 'ftm', 'fta', \
       'tpm','tpa', 'tpp', 'offReb', 'defReb', 'totReb', \
       'assists', 'pFouls','steals', 'turnovers', 'blocks', 'plusMinus']]

pos_df_aggregate = player_df_details_2.groupby(['pos']).agg('sum')
pos_df_aggregate['fgp']=pos_df_aggregate['fgm']/pos_df_aggregate['fga']
pos_df_aggregate['ftp']=pos_df_aggregate['ftm']/pos_df_aggregate['fta']
pos_df_aggregate.columns
```
