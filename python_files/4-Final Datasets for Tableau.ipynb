```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
player_df_details= pd.read_csv('Player_Details_for_Final_Tables.csv')
player_df_details.info()
player_df_details.columns
player_df_details.head()
```

```python
player_pca_clusters_df= pd.read_csv('player_pca_clusters.csv')
player_pca_clusters_df.reset_index(drop=True,inplace=True)
player_pca_clusters_df.set_index("player_id",inplace=True)

player_pca_clusters_df.info()
player_pca_clusters_df.columns
player_pca_clusters_df.head()
```

```python
player_df_details_merged=pd.merge(player_df_details,player_pca_clusters_df,on=["player_id"],how='inner')
```

```python
player_df_details_merged.shape
player_df_details_merged.columns
```

```python
player_df_details_1 = player_df_details_merged[['PlayerCluster','points', 'tpm','tpa','offReb', 'defReb', 'totReb', 'assists', 'pFouls', 'steals','turnovers', 'blocks', 'fga','fgm', 'fta','ftm']]

PlayerCluster_df_aggregate = player_df_details_1.groupby(['PlayerCluster']).agg('sum')
PlayerCluster_df_aggregate['fgp']=PlayerCluster_df_aggregate['fgm']/PlayerCluster_df_aggregate['fga']
PlayerCluster_df_aggregate['ftp']=PlayerCluster_df_aggregate['ftm']/PlayerCluster_df_aggregate['fta']
PlayerCluster_df_aggregate['tpp']=PlayerCluster_df_aggregate['tpm']/PlayerCluster_df_aggregate['tpa']
PlayerCluster_df_aggregate.head()
```

```python
player_df_details_2 = player_df_details_merged[['pos','points', 'tpm','tpa','offReb', 'defReb', 'totReb', 'assists', 'pFouls', 'steals','turnovers', 'blocks', 'fga','fgm', 'fta','ftm']]

pos_df_aggregate = player_df_details_2.groupby(['pos']).agg('sum')
pos_df_aggregate['fgp']=pos_df_aggregate['fgm']/pos_df_aggregate['fga']
pos_df_aggregate['ftp']=pos_df_aggregate['ftm']/pos_df_aggregate['fta']
pos_df_aggregate['tpp']=pos_df_aggregate['tpm']/pos_df_aggregate['tpa']
pos_df_aggregate.head()
```

```python
player_df_details_3 = player_df_details_merged[['player_id','firstname', 'lastname','points', 'tpm','tpa','offReb', 'defReb', 'totReb', 'assists', 'pFouls', 'steals','turnovers', 'blocks', 'fga','fgm', 'fta','ftm']]

Player_df_aggregate = player_df_details_3.groupby(['player_id','firstname', 'lastname']).agg('sum')
Player_df_aggregate['fgp']=Player_df_aggregate['fgm']/Player_df_aggregate['fga']
Player_df_aggregate['ftp']=Player_df_aggregate['ftm']/Player_df_aggregate['fta']
Player_df_aggregate['tpp']=Player_df_aggregate['tpm']/Player_df_aggregate['tpa']
Player_df_aggregate.head()
```

```python
# Exporting Actual metric data
PlayerCluster_df_aggregate_t=PlayerCluster_df_aggregate.transpose()
pos_df_aggregate_t=pos_df_aggregate.transpose()
Player_df_aggregate_t=Player_df_aggregate.transpose()

PlayerCluster_df_aggregate_t.to_json("Cluster_Stats_Aggregate")
pos_df_aggregate_t.to_json("Position_Stats_Aggregate")
Player_df_aggregate_t.to_json("Player_Stats_Aggregate")
```

```python
scaler = MinMaxScaler()

Player_aggregate_scaled = scaler.fit_transform(Player_df_aggregate[['points', 'tpm', 'tpa', 'offReb', 'defReb', 'totReb', 'assists',\
       'pFouls', 'steals', 'turnovers', 'blocks', 'fga', 'fgm', 'fta', 'ftm',\
       'fgp', 'ftp', 'tpp']])

# Create a DataFrame with the scaled data
Player_df_aggregate_scaled = pd.DataFrame(Player_aggregate_scaled,columns=['points', 'tpm', 'tpa', 'offReb', 'defReb', 'totReb', 'assists',\
       'pFouls', 'steals', 'turnovers', 'blocks', 'fga', 'fgm', 'fta', 'ftm',\
       'fgp', 'ftp', 'tpp'])


# Set the player_id column as index
Player_df_aggregate_scaled.index = Player_df_aggregate.index

# Display sample data
Player_df_aggregate_scaled.head()
```
